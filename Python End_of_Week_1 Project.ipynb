#### 1. Robust Calculator (types & conversions, error handling)

```python
print("What kind of operation would you like to perform?\n")

operator = input("Use (+) for Addition.\n(-) for Subtraction.\n(*) for Multiplication.\n(/) for Division of integers.\n")
num1 = int(input('Input first digit'))
num2 = int(input('Input second digit'))

if operator == '+':
    solution = (num1 + num2)
    print(solution)
elif operator == '-':
    solution = (num1 - num2)
    print(solution)
elif operator == '*':
    solution = (num1 * num2)
    print(solution)
elif operator == '/':
    solution = (num1 / num2)
    print(solution)


print(f"{num1} {operator} {num2} = {solution}")
```

#### 2. Student Records — CRUD with Lists & Dicts

```python
# Initial list to store student dictionaries 
students = []

# Example of adding a record
s_id = input("Enter Student ID: ")
name = input("Enter Name: ")
age = input("Enter Age: ")
gpa = input("Enter GPA: ")

# Validate age is numeric and GPA is 0.0–5.0
if age.isnumeric() and 0.0 <= float(gpa) <= 5.0:
    # Create the dictionary and add to list 
    new_student = {
        "id": s_id,
        "name": name,
        "age": int(age),
        "gpa": float(gpa)
    }
    students.append(new_student)
    print("Record saved!")
else:
    print("Invalid input: Age must be a number and GPA between 0.0-4.0.") [cite: 9]

# print listing of all students
print("Current Records:", students)
```

#### 3. Nested Grading Logic (control flow & aggregation)

```python
scores = [95, 82, 77, 91, 88] # Assessment scores 
grades = []

for score in scores:
    # Nested conditionals for letter grades and +/- tiers 
    if score >= 90:
        if score >= 97: letter = "A+"
        elif score >= 93: letter = "A"
        else: letter = "A-"
    elif score >= 80:
        if score >= 87: letter = "B+"
        elif score >= 83: letter = "B"
        else: letter = "B-"
    else:
        letter = "F"
    grades.append(letter)

# Calculate the mean (average)
mean_score = sum(scores) / len(scores)
print("Grade Distribution:", grades)
print("Class Mean:", mean_score)
```

#### 4. Contact Search & Deduplication (strings & comprehensions)

```python
contacts = [
    {"name": "Alice", "email": "alice@test.com"},
    {"name": "Bob", "email": "bob@test.com"}
]

new_name = "Alice"
new_email = "alice@test.com"

# Only add if email doesn't exist to remove duplicates and give error if there is
is_duplicate = False
for person in contacts:
    if person["email"].lower() == new_email.lower():
        is_duplicate = True
        print(f"Log: Removed duplicate for {new_email}") [cite: 15]

if not is_duplicate:
    contacts.append({"name": new_name, "email": new_email})
```

#### 5. Loop Practice — Prime Gap Finder (loops)

```python
last_prime = 2
# Loop up to 10,000
for num in range(3, 10001):
    # Check if number is prime
    is_p = True
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            is_p = False
            break
    
    if is_p:
        if (num - last_prime) >= 20:
            print(f"Gap found: {last_prime} and {num} (Gap: {num - last_prime})")
        last_prime = num
```

#### 6. BMI Logger with Unit Options (I/O, validation, datetime)

```python
import datetime

system = input("Choose system (metric/imperial): ").lower()
weight = float(input("Enter weight: "))
height = float(input("Enter height: "))

# Convert units and choose BMI
if system == "imperial":
    bmi = (weight / (height ** 2)) * 703
else:
    bmi = weight / (height ** 2)

# choose category
if bmi < 18.5: category = "Underweight"
elif bmi < 25: category = "Healthy"
else: category = "Overweight"

# Create a timestamped record
timestamp = datetime.datetime.now().strftime("%Y-%m-%d %H:%M")
print(f"Date: {timestamp} | BMI: {bmi:.1f} | Category: {category}")
```

#### 7. Adaptive Guessing Game (loops, state persistence)

```python
import random

streak = 0
playing = True

while playing:
    # Difficulty range adjusts based on player streaks 
    upper_limit = 10 + (streak * 10)
    secret = random.randint(1, upper_limit)
    
    print(f"Guess a number 1-{upper_limit}!")
    guess = int(input("Your guess: "))
    
    if guess == secret:
        print("Correct!")
        streak += 1 # Track the players streak
    else:
        print(f"Wrong! The number was {secret}.")
        streak = 0 # Reset the player streak to zero if he/she fail to guess the correct number
        playing = False # End game
```
